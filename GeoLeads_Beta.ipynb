```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from geopy.geocoders import Nominatim
import time

# 1. Configurando o Geocodificador (O motor que descobre as coordenadas)
# Usamos um 'user_agent' amigável para o servidor não bloquear a nossa busca de mapa
geolocator = Nominatim(user_agent="geoleads_mvp_context")

print("🗺️ Lendo os dados gerados pelo DuckDB...")

# Simulando a leitura do DataFrame de elite estruturado na etapa anterior
dados_leads = [
    {"CNPJ": "05.432.111/0001-90", "Empresa": "LOGISTICA ALPHAVILLE EXPRESS LTDA", "Capital": "R$ 1.200.000", "Telefone": "(11) 4191-0011", "Email": "diretoria@alphaexpress.com.br", "Bairro": "Tambore", "Cidade": "Barueri"},
    {"CNPJ": "10.987.654/0001-33", "Empresa": "MEDICAMENTOS OSASCO DISTRIBUIDORA S/A", "Capital": "R$ 4.500.000", "Telefone": "(11) 3685-2233", "Email": "financeiro@osascomed.com.br", "Bairro": "Industrial", "Cidade": "Osasco"},
    {"CNPJ": "20.123.456/0001-44", "Empresa": "PARNAIBA EMBALAGENS INDUSTRIAIS", "Capital": "R$ 800.000", "Telefone": "(11) 4154-8899", "Email": "rh@parnaibapack.com.br", "Bairro": "Fazendinha", "Cidade": "Santana de Parnaiba"},
    {"CNPJ": "08.765.432/0001-88", "Empresa": "ALFA SEGURANCA E SERVICOS PATRIMONIAIS", "Capital": "R$ 2.500.000", "Telefone": "(11) 4208-5566", "Email": "comercial@alfaseg.com.br", "Bairro": "Alphaville", "Cidade": "Barueri"},
    {"CNPJ": "15.432.999/0001-00", "Empresa": "TRANSPORTADORA ROCHDALE LTDA", "Capital": "R$ 600.000", "Telefone": "(11) 3651-7788", "Email": "adm@transrochdale.com.br", "Bairro": "Rochdale", "Cidade": "Osasco"}
]

df = pd.DataFrame(dados_leads)

# 2. Inicializando o Mapa centralizado na região oeste de São Paulo
# Coordenadas médias da nossa região alvo (Barueri/Osasco)
mapa = folium.Map(location=[-23.5109, -46.8183], zoom_start=12, tiles="OpenStreetMap")

# Criamos um agrupador de marcadores para o mapa ficar limpo e profissional
cluster = MarkerCluster().add_to(mapa)

print("📡 Convertendo endereços em coordenadas geográficas (Geocodificação)...")

for index, linha in df.iterrows():
    # Monta a string de busca para o mapa encontrar com precisão
    endereco_busca = f"{linha['Bairro']}, {linha['Cidade']}, SP, Brasil"
    print(f"📍 Mapeando: {linha['Empresa']} ({linha['Bairro']} - {linha['Cidade']})...")

    try:
        localizacao = geolocator.geocode(endereco_busca, timeout=10)
        if localizacao:
            # Texto em HTML estruturado que vai aparecer quando o usuário clicar no ponto do mapa
            html_popup = f"""
            <div style="font-family: Arial, sans-serif; width: 250px; font-size: 12px;">
                <h4 style="margin: 0 0 5px 0; color: #1a73e8;">{linha['Empresa']}</h4>
                <b>CNPJ:</b> {linha['CNPJ']}<br>
                <b>Capital Social:</b> <span style="color: green; font-weight: bold;">{linha['Capital']}</span><br>
                <hr style="border: 0; border-top: 1px solid #ccc; margin: 5px 0;">
                <b>📞 Telefone:</b> {linha['Telefone']}<br>
                <b>✉️ E-mail:</b> <a href="mailto:{linha['Email']}">{linha['Email']}</a><br>
                <b>📍 Local:</b> {linha['Bairro']}, {linha['Cidade']}
            </div>
            """
            iframe = folium.IFrame(html=html_popup, width=270, height=150)
            popup_final = folium.Popup(iframe, max_width=290)

            # Adiciona o alfinete no mapa dentro do cluster
            folium.Marker(
                location=[localizacao.latitude, localizacao.longitude],
                popup=popup_final,
                tooltip=f"Clique para ver detalhes de: {linha['Empresa']}",
                icon=folium.Icon(color="blue", icon="briefcase", prefix="fa")
            ).add_to(cluster)

        time.sleep(1.0) # Pausa obrigatória de segurança para respeitar o servidor do mapa
    except Exception as e:
        print(f"⚠️ Não foi possível mapear a empresa {linha['Empresa']}: {e}")

print("\n🎯 Mapa gerado com sucesso!")
print("👇 Veja o resultado interativo renderizado logo abaixo:")
mapa
```
